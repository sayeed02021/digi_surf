```python
def parse_args():
    with open('saved_stuff/configs.yaml', 'r') as f:
        config_data = yaml.safe_load(f)

    args = argparse.Namespace(**config_data)
    configs_save_path = f'{args.save_folder}/configs.yaml'
    os.makedirs(args.save_folder, exist_ok=True)
    with open(configs_save_path, 'w') as f:
        yaml.safe_dump(config_data, f)
    return args
```

```python
args = parse_args()
os.makedirs(args.save_folder, exist_ok=True)
tokenizer = Tokenizer()
tokenizer.load_vocab(args.vocab_path)

if not torch.cuda.is_available():
    args.device = 'cpu'

df_train = pd.read_csv('data/train.csv')
train_data,  val_data = get_split_dataset(
        df = df_train,
        tokenizer=tokenizer,
        max_seq_len=args.max_seq_len,
        args = args
    )


df_test = pd.read_csv('data/test.csv')

test_data = OptDataset(
    src_smiles = df_test['Source_Mol'].to_numpy(),
    src_prop = df_test['Source_Mol_pCMC'].to_numpy(),
    tgt_smiles = df_test['Target_Mol'].to_numpy(),
    tgt_prop = df_test['Target_Mol_pCMC'].to_numpy(),
    tokenizer=tokenizer,
    max_seq_len=args.max_seq_len
)


print("Length of train, val, test: ", len(train_data), len(val_data), len(test_data))

train_loader = DataLoader(train_data, batch_size=args.batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=args.batch_size, shuffle=False)
test_loader = DataLoader(test_data, batch_size=20, shuffle=False)
```

```python
def save_generated_data(generated_data, loader, tokenizer):
    data = defaultdict(list)
    for data_batch, gen_batch in zip(loader, generated_data):
        src, tgt, p = data_batch 
        tgt2, gen_smiles_per_batch = gen_batch 

        tgt = [tokenizer.decode(t.cpu().tolist()) for t in tgt]
        src = [tokenizer.decode(s.cpu().tolist()) for s in src]
        
        tgt_prop = p[:,0]
        src_prop = p[:,1]
        

        for gen_smiles_per_tgt, src_smile, tgt_smile, prop in zip(gen_smiles_per_batch, src, tgt, p):
            data['Source_Mol'].append(src_smile)
            data['Target_Mol'].append(tgt_smile)
            data['Source_Mol_pCMC'].append(prop[1].item())
            data['Target_Mol_pCMC'].append(prop[0].item())
            for idx, gen_smi in enumerate(gen_smiles_per_tgt):
                data[f'Predicted_smi_{idx+1}'].append(gen_smi)

    df = pd.DataFrame(data)
    return df


```

```python
df = save_generated_data(sequences, test_loader, tokenizer)
```
